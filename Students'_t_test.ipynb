**Student's t-test**

1- One-sample t-test\
2- Two-sample t-test\
    i- un-paired or indepenent t-test\
    ii- paired or relational/ dependent t-test

**One sample t-test**

Test a sample with a non standard value.

**Assumptions**\
    1- Observation in sample is independent and identically distributed.\
    2- Observation in sample is normally distributed.

**Interpretation**\
    **H0:** mean of sample is equal to the known value.\
    **H1:** mean of sample is unequal to the known value.

**Import libraries**

```python
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_1samp
```

Load Dataset

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df1= df[['age', 'sex']]
df1.head()
```

```python
df1= df[['age', 'sex','fare']]
df1.head()
```

Data description

```python
df1.describe()
```

Note: std shows data is not normal

Check the age and compare with a known value of 45 years

```python
ttest_1samp(df1['age'], 45)
```

```python
# Compare with 50 years age
ttest_1samp(df1['age'], 50)
stat, p= ttest_1samp(df1['age'], 50)
print('stat= %.3f, p %.3f' % (stat, p))

# Adding conditional argument for ease
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distribution')
```

```python
ttest_1samp(df1['fare'], 50)
stat, p= ttest_1samp(df1['fare'], 50)
print('stat= %.3f, p %.3f' % (stat, p))

# Adding conditional argument for ease
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distribution')
```

**Note:** Std is more than mean in the fare column, shows a flatten kurtosis.

**Two sample t-test**

i- **Independent student's t-test**

**Assumptions**\
    1- Observations in sample is independent and identically distributed.\
    2- Observations in sample is normally distributed.\
    3- Observations in sample have same variance.

**Interpretation**\
    **H0:** mean of samples are equal.\
    **H1:** mean of samples areunequal.

```python
#Now we will compare age of male and female

# 1- Split the dataset
df_male = df1.loc[df1['sex']== 'male']
df_female = df1.loc[df1['sex']== 'female']
```

Import Libraries

```python
from scipy.stats import ttest_ind

stat, p= ttest_ind(df_male['age'], df_female['age'])

print('stat= %.3f, p %.3f' % (stat, p))

# Adding conditional argument for ease
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distribution')
```

ii- **Dependent/ Paired student's t-test**

**Assumptions**\
    1- Observations in sample is independent and identically distributed.\
    2- Observations in sample is normally distributed.\
    3- Observations in sample have same variance.\
    4- Observations across each samples are paired.

**Interpretation**\
    **H0:** mean of samples are equal.\
    **H1:** mean of samples are unequal.
